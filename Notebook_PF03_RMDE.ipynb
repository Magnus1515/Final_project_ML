```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import ast
import itertools
# pd.set_option('display.max_columns', None)
```

```python
# PREPROCESAMIENTO DE DATOS
bb = pd.read_csv('https://raw.githubusercontent.com/Magnus1515/PruebaGithub/master/Best_Book_21st.csv')
bb = bb.fillna(0)
bb = bb.drop_duplicates(subset='title')
bb.head(2)
```

```python
# Se da un input de entrenamiento poniendo titulos y asignando ratings (imitando la preferencia de un usuario)
inputUser = [
            {'title':'The Glass Castle', 'rating':1},
            {'title':'Catching Fire (The Hunger Games)', 'rating':2},
            {'title':'Harry Potter and the Cursed Child: Parts One and Two (Harry Potter, #8) ', 'rating':3},
            {'title':'The Martian', 'rating':3},
            {'title':'The Life of Pi', 'rating':5}
         ] 
inputBooks = pd.DataFrame(inputUser)
inputBooks
```

```python
# busca el libro y la asigna si id y genero
inputID = bb[bb['title'].isin(inputBooks['title'].tolist())]
inputBooks = pd.merge(inputID, inputBooks)
inputBooks.drop(columns = ['book_link','series','date_published','num_of_page',
                           'publisher','award','rating_count', 'review_count',
                           'rating_count','rate','lang'], inplace = True )
inputBooks = inputBooks.drop_duplicates(subset='id')
inputBooks
```

```python
# se remueven columnas innecesarias, y se hace el metodo bag of letters, separando todas las letras de todos los generos, y al pasar por el genero 
# de cierto libro marca con uno si ese genero tiene una letra y 0 sino. Eso se hace para todos los libros del dataset

BooksWithGenres_df = bb.copy()

for index, row in bb.iterrows():
    if row['genre'] == 0:
      continue
    for genre in row['genre']:
        BooksWithGenres_df.at[index, genre] = 1

BooksWithGenres_df = BooksWithGenres_df.fillna(0)
BooksWithGenres_df.drop(columns = ['book_link','series','date_published','num_of_page',
                           'publisher','award','rating_count', 'review_count',
                           'rating_count','rate','lang'], inplace = True )
BooksWithGenres_df.head(2)
```

```python
# separa los generos que estan divididos por comas
BooksWithGenres_df['genre'] = BooksWithGenres_df.genre.str.split(',')
BooksWithGenres_df
```

```python
# despues se busca en esos titulos de bag of letters los libros input del usuario para obtener sus bag of letters
userBooks = BooksWithGenres_df[BooksWithGenres_df['id'].isin(inputBooks['id'].tolist())]
userBooks
```

```python
# Se remueven las columnas

userBooks = userBooks.reset_index(drop = True)
userGenreTable = userBooks.drop(columns = ['id', 'title','author','genre'])
userGenreTable
```

```python
# obtiene los ratings del libro input

inputBooks['rating']
```

```python
# obtiene los scores a traves de hacer un producto punto con la lista de ratings y la matriz de bag of letters de cada libro
userProfile = userGenreTable.transpose().dot(inputBooks['rating'])
userProfile
```

```python
# remueve columnas innecesarias y pone los ids a los libros
genreTable = BooksWithGenres_df.set_index(BooksWithGenres_df['id'])
genreTable = genreTable.drop(columns = ['id','title','author','genre'])
genreTable.head
```

```python
# hace el calculo del porcentaje de similitud entre los libros de input y los del dataset: multiplica los scores de los inputs y la matriz de bag of letters y los suma para cada cada uno, al final divide esa cantidad 
# entre la suma de los scores del input
recommendationTableDF = ((genreTable * userProfile).sum(axis=1))/(userProfile.sum())
recommendationTableDF.head()
```

```python
# al hacer ordaniento de mayor a menor, se obtienen lo5 libros con el porcentaje mas alto de probabilidad de coincidencia, y por consiguiente, las recomendaciones
recommendationTableDF = recommendationTableDF.sort_values(ascending = False)
recommendationTableDF.head(50)
```

```python
# Se buscan los datos de los libros recomendados 
bb = bb.loc[bb['id'].isin(recommendationTableDF.head(5).keys())]
bb = bb.drop(columns = ['book_link','publisher','num_of_page', 'lang',
                    'review_count', 'rating_count', 'rate', 'award'])
bb.drop(columns = ['series'])
```
